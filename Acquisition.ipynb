```python
from selenium import webdriver
from time import sleep
global driver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup  
import pandas as pd
import numpy as np
import requests
```

```python
#Lists declarations
name = list()
age = list()
overall = list()
potential = list()
position = list()
value = list()
pace = list()
shooting = list()
passing = list()
dribbling = list()
physical = list()
defending = list()
total_stats = list()
```

```python
#Preparing the fixed url for the data acquisition by SELENIUM
def get_url():
    options = Options()
    options.add_argument("--headless")
    options.add_argument("window-size=1920,1080")

    sofifa_Url = "https://sofifa.com/"

    driver = webdriver.Chrome(r'C:\Users\Niv\Desktop\Stuff\HIT\Data Science/chromedriver.exe',options=options)
    driver.get(sofifa_Url)
    action = webdriver.ActionChains(driver)
    driver.find_element_by_xpath('//*[@id="column-selected"]/form/div[1]/div/div[1]/input').click() #col_select
    sleep(1)
    #choose attribute and apply them
    driver.find_element_by_xpath("//div[@data-id='9']").click()
    driver.find_element_by_xpath("//div[@data-id='64']").click()
    driver.find_element_by_xpath("//div[@data-id='65']").click() 
    driver.find_element_by_xpath("//div[@data-id='66']").click()
    driver.find_element_by_xpath("//div[@data-id='67']").click()
    driver.find_element_by_xpath("//div[@data-id='68']").click()
    driver.find_element_by_xpath("//div[@data-id='69']").click()
    driver.find_element_by_xpath('//*[@id="body"]/header/div[3]/div/form/input').click()
    driver.find_element_by_xpath('//*[@id="column-selected"]/form/div[2]/button').click()


    #base for BS
    Current_Url = driver.current_url
    return Current_Url
```

```python
#Loading bs object 
def load_soup(url):
    r = requests.get(url)
    return BeautifulSoup(r.content, "html.parser")
```

```python
#Crawling SOFIFA
#crawling with BS

def create_df(page_number, url):
    link = url
    n_pages = page_number
    i = 0


    for i in range (n_pages):
    
        soup = load_soup(link)
        tbl = soup("tbody",attrs = {'class':'list'})[0]
    
        for row in tbl("tr"):
            cells = row("td")
            name.append(cells[1]("a")[0]("div")[0].text)
            age.append(cells[2].text)
            overall.append(cells[3]("span")[0].text)
            potential.append(cells[4]("span")[0].text)
            position.append(cells[6]("a")[0]("span")[0].text)
            value.append(cells[7].text)
            total_stats.append(cells[9]("span")[0].text)
            pace.append(cells[10]("span")[0].text)
            shooting.append(cells[11]("span")[0].text)
            passing.append(cells[12]("span")[0].text)
            dribbling.append(cells[13]("span")[0].text)
            defending.append(cells[14]("span")[0].text)
            physical.append(cells[15]("span")[0].text)
        
        
        
        
        if(i == 0):
            link = 'https://sofifa.com/' + soup("div",attrs = {"class":"pagination"})[0]('a')[0].get("href") 
        else:
            link = 'https://sofifa.com/' + soup("div",attrs = {"class":"pagination"})[0]('a')[1].get("href") 
   
    df=pd.DataFrame({"name":name,"age":age,"overall":overall,"potential":potential,"position":position,"pace":pace,"shooting":shooting,"passing":passing,"dribbling":dribbling,"defending":defending,"physical":physical,"total_stats":total_stats,"value":value})
    return df
```

```python
Current_url = get_url()
df= create_df(120,Current_url)
df
```
